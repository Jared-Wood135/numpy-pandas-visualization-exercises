## Make a scatterplot
- Flipper length and mass
- Broken down by penguin species

    Gentoo
    
    Chinstrap
    
    Adelie

```python
gentoo = penguins[penguins['species'] == 'Gentoo']
adelie = penguins[penguins['species'] == 'Adelie']
chinstrap = penguins[penguins['species'] == 'Chinstrap']
```

```python
plt.scatter(gentoo.flipper_length_mm, gentoo.body_mass_g, label='gentoo')
plt.scatter(adelie.flipper_length_mm, adelie.body_mass_g, label='adelie')
plt.scatter(chinstrap.flipper_length_mm, chinstrap.body_mass_g, label='chinstrap')
plt.legend();
```

### Pie Chart

- Proportion of data
- Uses frequency

```python
titanic = data('titanic')
```

```python
titanic.info()
```

```python
titanic.sample(5)
```

```python
survivors = sum(titanic.survived == 'yes')
non_survivors = sum(titanic.survived == 'no')
```

```python
survivors, non_survivors
```

```python
# Assign labels
my_labels = ['Survived', 'Did Not Survive']

# Set the size of each part
sizes = [survivors, non_survivors]

# Create the figure
# Create the subplot (ax)

fig, ax = plt.subplots()
ax.pie(sizes, labels = my_labels, autopct = '%1.0f%%', colors = ('tan', 'firebrick'))

ax.set_title('Survival Rate  for  Titanic');
```
